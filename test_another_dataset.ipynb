```python
import pandas as pd
```

```python
df = pd.read_excel("air_quality_data_edited.xlsx")


df = df.drop(df.columns[0], axis=1)
df = df.rename(columns={'Core Based Statistical Area': 'city'})
df = df.iloc[:1018][['city', 'Pollutant', 2020]]

df['city'] = df['city'].fillna(method='ffill')

new_df = df.pivot_table(
    index='city',
    columns='Pollutant',
    values=2020,
    aggfunc='first'
)

new_df = new_df.reset_index()

new_df
```
